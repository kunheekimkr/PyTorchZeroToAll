# Lecture 03: Gradient Descent

## What is the Learning?

- Finding Coefficients that minimizes the Loss!

But how?

## Gradient Descent Algorithm
$$ gradient = \frac{\partial loss}{\partial w} $$

If gradient>0, minimum is on the right

If gradient<0, minimum is on the left.

$$ w = w - \alpha \frac{\partial loss}{\partial w} =  w - \alpha * 2x(xw-y) $$


## Data, Model, Loss, and Gradient

```python
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]
w = 1.0 # a random guess: random value

#our model for the forward pass
def forward(x):
    return x*w

#loss function
def loss(x,y):
    y_pred = forward(x)
    return (y_pred - y) * (y_pred - y)

#compute gradient
def gradient (x,y):
    return 2* x * (x*w -y)
```

## Training: update weight

```python
print("predict (before training)", 4, round(forward(4),3))

# Training Loop
for epoch in range(100):
    for x_val, y_val in zip(x_data, y_data):
        grad= gradient(x_val, y_val)
        w = w -0.01 * grad
        print('\tgrad: ', x_val, y_val, round(grad,3))
        l = loss(x_val, y_val)

    print("progress:", epoch, "w=", round(w,1), "loss=", round(l,3))

# After Training
print("predict (after training)", "4", round(forward(4),3))
```
